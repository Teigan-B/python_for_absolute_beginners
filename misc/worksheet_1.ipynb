```python
ethnicity_language_url = "https://raw.githubusercontent.com/data-to-insight/D2I-Jupyter-Notebook-Tools/main/DfE_public_data/spc_pupils_ethnicity_and_language_.csv"

EthLang = pd.read_csv(ethnicity_language_url)

EthLang202122ESCC = EthLang[(EthLang["la_name"] == "East Sussex") & (EthLang["time_period"] == 202122)]

print(EthLang202122ESCC)
```

```python
EthLang202122ESCC_ethnicity = EthLang202122ESCC[(EthLang202122ESCC["phase_type_grouping"] == "Total") 
                                                & ~(EthLang202122ESCC["ethnicity"] == "Total")]

print(EthLang202122ESCC_ethnicity[["ethnicity","headcount"]])
```

```python
EthLang202122Bromley = EthLang[(EthLang["la_name"] == "Bromley") & (EthLang["time_period"] == 202122)]

print(EthLang202122Bromley)
```

```python
EthLangBromley_ethnicity = EthLang202122Bromley[(EthLang202122Bromley["phase_type_grouping"] == "Total")
                                                & ~(EthLang202122Bromley["ethnicity"] == "Total")]

print(EthLangBromley_ethnicity[["la_name", "ethnicity", "headcount"]])
```

```python
EthLangBromley_ethnicity_columns = EthLangBromley_ethnicity[["la_name", "ethnicity", "headcount"]]

print(EthLangBromley_ethnicity_columns)
```

```python
print("The total number of people, regardless of ethnicities is:")
print(EthLangBromley_ethnicity_columns["headcount"].sum())

print("The average number of people for all ethnicities is:")
print(EthLangBromley_ethnicity_columns["headcount"].mean())

print(EthLangBromley_ethnicity_columns["headcount"].min())
print(EthLangBromley_ethnicity_columns["headcount"].max())

# if EthLangBromley_ethnicity_columns["headcount"] == EthLangBromley_ethnicity_columns["headcount".min()]:
  #   print(EthLangBromley_ethnicity_columns["ethnicity"])

print("The minumum number of people represented in any ethnicity is:")
print(EthLangBromley_ethnicity_columns[EthLangBromley_ethnicity_columns["headcount"] == (EthLangBromley_ethnicity_columns["headcount"].min())])

print("The maximum number of people represented in any ethnicity is:")
print(EthLangBromley_ethnicity_columns[EthLangBromley_ethnicity_columns["headcount"] == EthLangBromley_ethnicity_columns["headcount"].max()])
```

```python
fig_ethnicity = sns.barplot(data=EthLangBromley_ethnicity_columns, x = "ethnicity", y = "headcount", color = "red")

plt.xticks(rotation = 90)

fig_ethnicity.set_ylabel("Headcount")
fig_ethnicity.set_xlabel("Ethnicity Subgroups")
fig_ethnicity.set_title("Bromley 2021/22 Ethnicity Breakdown")

plt.show()
```
