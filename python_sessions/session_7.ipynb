```python
# make a dictionry of 5 names and job titles
# use k,v loop to print something like "My name is NAME and I am a JOB TITLE"
# make a new dicitonary of names as they keys and the new f strings as the values

name_job = {"Andrew":"accountant", 
            "Ben":"builder", 
            "Chad":"clown", 
            "Derek":"designer", 
            "Fido":"dog"}

name_phrase = {}

for name, job in name_job.items():
    name_job_phrase = f"My name is {name} and my job is {job}."
    print(name_job_phrase)
    name_phrase[name] = name_job_phrase

name_phrase
```

```python
# make a dictionry of 5 names and job titles
# use k,v loop to print something like "My name is NAME and I am a JOB TITLE"
# make a new dicitonary of names as they keys and the new f strings as the values
 
name_job = {'Will':'Data manager',
            'Annie':'Retired',
            'Andy':'Retired',
            'Naiomi':'Policy Writer',
            'Scout':'Moral Support'}
 
name_phrase = {}
 
for k, v in name_job.items():
    if (k == 'Annie') | (k == 'Andy'):
        phrase = f'My name is {k}, and I am {v}'
    else:
        phrase = f'My name is {k}, and I am a {v}'
    name_phrase[k] = phrase
 
name_phrase
```

```python
# nested loops: go through all the loops in the first list first
evens = list(range(2,13,2))
odds = list(range(1,13,2))

multiply_list = []

for i in evens:
    print(f"The current i is {i}")
    for j in odds:
        print(i * j)

for i in evens:
    print(f"The current i is {i}")
    for j in odds:
        multiply_list.append(i*j)
    print(multiply_list)
```

```python
# using these two lists of sandwich ingredients, find (using for loops) all of the spread/topping combinations
# append them to a new list of all the combos
spreads = ['butter', 'jam', 'peanut butter']
toppings = ['ham', 'cheese', 'sardines']

combo_list = []

for spread in spreads:
    for topping in toppings:
        combo = f"{spread} & {topping}"
        print(combo)
        combo_list.append(combo) # extend would extend by every indexable element of each thing so would be b,u,t,t,e,r
print(combo_list)
```

```python
# fizzbang
# take all the numbers from 1-100,
# print them, BUT if the number is divisible by 3, print fizz,
# if it's divisible by 5 print bang
# BUT if it's divisible by 3 & 5 print fizzbang

numbers_to_100 = list(range(1,101))
fizzbang = []

for i in numbers_to_100:
    if (i%5 == 0) & (i%3 == 0):
        i = "fizzbang"
    elif (i%5 == 0):
        i = "bang"
    elif (i%3 == 0):
        i = "fizz"
    fizzbang.append(i)

print(fizzbang)

numbers = list(range(1,101))
 
for number in numbers:
    if (number % 3 == 0) & (number % 5 ==0):
        print('fizzbang')
    elif (number % 3 == 0):
        print('fizz')
    elif (number % 5 == 0):
        print('bang')
    else:
        print(number)
 
for number in numbers:
    if (number % 3 == 0):
        if (number % 5 ==0):
            print('fizzbang')
        else:
            print('fizz')
    elif (number % 5 == 0):
        print('bang')
    else:
        print(number)


```
