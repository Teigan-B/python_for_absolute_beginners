```python
# write a function that returns x / y and x - y,
# # always dividing the biggest by the smallest number
# # and always taking the smallest fromt he biggest


def divide_xy(x, y):
    if x > y:
        div = x/y
        sub = x - y
    else:
        div = y/x
        sub = y - x
    return div, sub

divide, subtract = divide_xy(24,4)
print(divide)
print(subtract)

# or 

def minus_divide(x,y):
    if x >= y:
        big = x
        small = y
    else:
        big = y
        small = x
    subtract = big - small
    division = big / small
 
    return subtract, division
 
sub, div = minus_divide(70, 100)
 
print(sub, div)

```

```python
# In this cell, first, import the 'math' module.
# Next, write a function that squares a number (hint: you can simply uses x*x),
# Now write a function that uses math.sqrt() to find the square root of a number
# (hint: https://docs.python.org/3/library/math.html#math.sqrt)
# Then, write a funciton that performs both of these functions and returns both answers,
# Finally, call the function getting two variables as your answer, and print them.

import math

def square(x):
    square_x = x**2
    return square_x

def sqrt(x):
    root_x = math.sqrt(x)
    return root_x

def square_and_root(x):
    root_x = sqrt(x)
    square_x = square(x)
    return root_x, square_x

print(square_and_root(9))

```

```python
'''
In child identifiers table, write a function that calculates ages based on the data, and returns a new column in age range buckets
0-5, 6-11, 12-17, 18+
convert value to a datetime
find the difference between the datetime and today
write if else statements to put that difference in time in a bucket
return as a string the appropriate buckets to a new column
'''

ci = pd.read_csv(r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv")
ci

def age_group(date):
    bd = pd.to_datetime(date, format = "%Y-%m-%d", errors = "coerce")
    age = pd.to_datetime("today") - bd
    age = age / pd.Timedelta(365.25, "d")
    if age <= 5:
        return "0 - 5"
    elif age <= 11:
        return "6 - 11"
    elif age <= 17:
        return "12 - 17"
    else:
        return "18+"

ci["Age Group"] = ci["PersonBirthDate"].apply(age_group)
ci[["PersonBirthDate", "Age Group"]]
```
